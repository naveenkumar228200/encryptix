```python
data['Age'] = data['Age'].fillna(data['Age'].mean());
```

```python
data['Embarked'] = data['Embarked'].fillna('S')
```

```python
sex_mapping = {'male' : 0, 'female' : 1};
data['Sex'] = data['Sex'].map(sex_mapping);
```

```python
E_mapping = {'S':0,'C':1,'Q':2};
data['Embarked'] = data['Embarked'].map(E_mapping);
```

```python
print(data.isnull().sum());
```

```python
sea.relplot(x = 'Age', y= 'Survived',data = data);
```

```python
sea.relplot(x = 'Fare', y= 'Survived',data = data);
```

```python
data.drop(columns = ['Cabin'],inplace = True)
```

```python
data['Family_size'] = data['SibSp']+data['Parch']+1;
data['Age*class'] = data['Age']+data['Pclass'];
```

```python
data.drop(columns = ['PassengerId','Name','Ticket'],inplace = True);
```

```python
from sklearn.linear_model import LogisticRegression;
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split;
```

```python
train = data.drop(['Survived'],axis = 1);
test = data['Survived'];
```

```python
X_train,x_test,Y_train,Y_test = train_test_split(train,test,test_size = 0.3,random_state = 42);
```

```python
regr = LogisticRegression(max_iter = 1000);
```

```python
regr.fit(X_train,Y_train);
```

```python
pred = regr.predict(x_test);
```

```python
accuracy = accuracy_score(Y_test,pred);
```
